```python
import pandas as pd
import seaborn as sns

from sklearn.preprocessing import StandardScaler
```

откроем train_df.csv

```python
train_df = pd.read_csv("../data/train_df.csv")
test_df = pd.read_csv("../data/test_df.csv")
train_df
```

```python
features = [f"feature_{num}" for num in range(0, 79)]
```

```python
X_train, y_train = train_df[features], train_df["target"]
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
```

```python
def check_const_features(df, feature_name: str, threshold: float = 0.1) -> bool:
    return df[feature_name].var() <= threshold
```

```python
const_features = []

for feature in features:
    if check_const_features(train_df, feature):
        const_features.append(feature)

len(const_features)
```

```python
keep_columns = list(set(features) - set(const_features))
keep_columns.append("target")

custom_train_df = train_df[keep_columns]
custom_test_df = test_df[keep_columns] 

custom_train_df.to_csv("../data/custom_train_df.csv", index=False)
custom_test_df.to_csv("../data/custom_test_df.csv", index=False)
```
